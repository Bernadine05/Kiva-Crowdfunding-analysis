### Converting "posted_time" to datetime

```python
kiva_loans['posted_time'] = pd.to_datetime(kiva_loans['posted_time'])
```

## Analysis of the loan growth rate metric

```python
total_monthly_loans = kiva_loans.groupby(kiva_loans['posted_time'].dt.to_period('M'))['id'].count() #Groupby month and count the number of loans
total_monthly_loans.index = total_monthly_loans.index.strftime('%b %Y') #Format the string to Year-Month format

plt.figure(figsize=(20,10))
sns.lineplot(x=total_monthly_loans.index, y=total_monthly_loans)
plt.xlabel('Month & Year')
plt.ylabel('Loan Count')
plt.title(' TOTAL AMOUNT OF LOANS PER MONTH (2014-2017)')
plt.xticks(rotation=45)  #Rotating x-axis labels for better readerbility
pd.option_context('mode.use_inf_as_na', True)
```
